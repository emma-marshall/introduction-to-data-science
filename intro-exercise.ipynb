```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# This data comes from the UCI ML repository:
# https://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset
# It is the daily number of users from a bike share program
df = pd.read_csv('day.csv')
```

```python
# shows a preview of the data
df.head()
```

```python
# shows some basic stats of the data
df.describe()
```

```python
# mean - hourly data counts (the 'cnt' column)
df.cnt.mean()
```

```python
# standard deviation - hourly data counts (the 'cnt' column)
df.cnt.std()
```

```python
# plot the counts ('cnt' column)
plt.plot(df.cnt)                                       # create a plot of 'cnt' column
plt.xlabel("Instant")                                  # label x axis
plt.ylabel("Count")                                    # label y axis
plt.title("Count of Users from Bike Share Program")    # title the plot
plt.show()
```

```python
#get max values, grouped by month
monthGrouped = df.groupby('mnth')
maxGroupedValues = monthGrouped.max()

print(maxGroupedValues)    
```
